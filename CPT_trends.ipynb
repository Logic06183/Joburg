```python
import ee
import geemap
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Initialize Earth Engine
ee.Initialize()

# Function to get climate data from ERA5
def get_climate_data(start_date, end_date, location):
    # Filter ERA5 dataset and select necessary bands
    era5_hourly = ee.ImageCollection('ECMWF/ERA5/MONTHLY') \
        .filterDate(start_date, end_date) \
        .select(['mean_2m_air_temperature', 'dewpoint_2m_temperature'])

    # Extract data for the specified location
    def extract_data(image):
        stats = image.reduceRegion(
            reducer=ee.Reducer.mean(),
            geometry=location,
            scale=30000
        )
        temp = stats.get('mean_2m_air_temperature')
        dewpoint = stats.get('dewpoint_2m_temperature')
        
        # Calculate relative humidity
        if temp is not None and dewpoint is not None:
            temp_c = ee.Number(temp).subtract(273.15)  # Convert temperature to Celsius
            dewpoint_c = ee.Number(dewpoint).subtract(273.15)  # Convert dewpoint to Celsius
            rh = ee.Number(100).multiply(
                ee.Number(17.625).multiply(dewpoint_c).divide(dewpoint_c.add(243.04))
                .subtract(
                    ee.Number(17.625).multiply(temp_c).divide(temp_c.add(243.04))
                ).exp()
            )
        else:
            rh = None

        return ee.Feature(None, {
            'temperature': temp,
            'humidity': rh,
            'date': image.date().format('YYYY-MM')
        })

    climate_data = era5_hourly.map(extract_data).getInfo()
    return climate_data

# Define location for Cape Town (coordinates: 18.4241° E, 33.9249° S)
cape_town = ee.Geometry.Point(18.4241, -33.9249)

# Get climate data for your study period
start_date = '2013-01-01'
end_date = '2019-12-31'
climate_data = get_climate_data(start_date, end_date, cape_town)

# Convert to pandas DataFrame
climate_df = pd.DataFrame([
    {
        'Month': feature['properties']['date'],
        'Temperature': feature['properties']['temperature'] - 273.15,  # Convert Kelvin to Celsius
        'Humidity': feature['properties']['humidity']
    }
    for feature in climate_data['features']
])

# Merge with your screening data
df['Month'] = pd.to_datetime(df['Month'])
climate_df['Month'] = pd.to_datetime(climate_df['Month'])

merged_df = pd.merge(df, climate_df, on='Month')

# Create visualization
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 18))

# Plot 1: Temperature and Screening Numbers
ax1.plot(merged_df['Month'], merged_df['Temperature'], color='red', label='Temperature (°C)')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

ax1_twin = ax1.twinx()
ax1_twin.plot(merged_df['Month'], merged_df['Screened'], color='blue', label='Screened Patients')
ax1_twin.set_ylabel('Number of Patients', color='blue')
ax1_twin.tick_params(axis='y', labelcolor='blue')

ax1.set_title('Cape Town: Temperature vs Screening Numbers')
ax1.grid(True)

# Plot 2: Humidity and Screening Numbers
ax2.plot(merged_df['Month'], merged_df['Humidity'], color='green', label='Relative Humidity (%)')
ax2.set_ylabel('Relative Humidity (%)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

ax2_twin = ax2.twinx()
ax2_twin.plot(merged_df['Month'], merged_df['Screened'], color='blue', label='Screened Patients')
ax2_twin.set_ylabel('Number of Patients', color='blue')
ax2_twin.tick_params(axis='y', labelcolor='blue')

ax2.set_title('Cape Town: Humidity vs Screening Numbers')
ax2.grid(True)

# Plot 3: Correlation Analysis
ax3.scatter(merged_df['Temperature'], merged_df['Screened'], c='red', alpha=0.5, label='Temperature')
ax3.scatter(merged_df['Humidity'], merged_df['Screened'], c='green', alpha=0.5, label='Humidity')
ax3.set_xlabel('Climate Variable (Temperature °C / Humidity %)')
ax3.set_ylabel('Number of Screened Patients')
ax3.set_title('Cape Town: Correlation between Climate Variables and Screening Numbers')
ax3.legend()
ax3.grid(True)

plt.tight_layout()
plt.show()

# Calculate correlations
correlations = merged_df[['Screened', 'Temperature', 'Humidity']].corr()
print("\nCorrelation Analysis for Cape Town:")
print(correlations)

# Seasonal analysis for Southern Hemisphere
merged_df['Season'] = merged_df['Month'].dt.month.map({
    12: 'Summer', 1: 'Summer', 2: 'Summer',  # Southern Hemisphere summer
    3: 'Autumn', 4: 'Autumn', 5: 'Autumn',
    6: 'Winter', 7: 'Winter', 8: 'Winter',
    9: 'Spring', 10: 'Spring', 11: 'Spring'
})

seasonal_stats = merged_df.groupby('Season').agg({
    'Screened': ['mean', 'std'],
    'Temperature': ['mean', 'std'],
    'Humidity': ['mean', 'std']
}).round(2)

print("\nSeasonal Analysis for Cape Town:")
print(seasonal_stats)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import calendar

# --------------------------------------------
# Prepare Data
# --------------------------------------------
df['Year'] = df['Month'].dt.year
df['MonthNum'] = df['Month'].dt.month
df['MonthName'] = df['MonthNum'].apply(lambda x: calendar.month_name[x])

# Exclude outliers (e.g., outside 5th and 95th percentile)
low_cut = df['Percentage'].quantile(0.05)
high_cut = df['Percentage'].quantile(0.95)
filtered_df = df[(df['Percentage'] >= low_cut) & (df['Percentage'] <= high_cut)]

# Compute a climate metric for each month if available (e.g., mean monthly temperature)
climate_stats = df.groupby('MonthNum')['Temperature'].mean().reset_index().rename(columns={'Temperature':'MeanTemp'})
filtered_df = filtered_df.merge(climate_stats, on='MonthNum', how='left')

# Order months from January to December
month_order = list(range(1, 13))

# --------------------------------------------
# Plotting
# --------------------------------------------
sns.set_theme(style="whitegrid")
plt.rcParams.update({
    'font.size': 12,
    'axes.titlesize': 14,
    'axes.labelsize': 12,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 10
})

fig, ax = plt.subplots(figsize=(10, 8))

# Vertical offset between ridges
y_off = 1.2  

cmap = sns.cubehelix_palette(n_colors=12, start=0.5, rot=0.75, gamma=0.8, dark=0.2, light=0.8, as_cmap=True)

for i, month in enumerate(month_order):
    month_data = filtered_df[filtered_df['MonthNum'] == month]['Percentage']
    if len(month_data) < 2:
        continue
    
    # Plot a temporary KDE to get the line coordinates
    temp_kde = sns.kdeplot(data=month_data, bw_adjust=0.5, cut=0, alpha=0)
    lines = temp_kde.get_lines()
    if not lines:
        # Remove the empty plot and continue
        temp_kde.collections.clear()
        continue

    x_coords = lines[0].get_xdata()
    y_coords = lines[0].get_ydata()

    # Clear the temporary KDE
    for coll in temp_kde.collections:
        coll.remove()
    for line in temp_kde.get_lines():
        line.remove()

    # Normalize each month's peak and scale it
    # This ensures all distributions are visible and not too flat.
    # For example, making the max height = 0.5 units above the baseline.
    max_y = y_coords.max()
    if max_y > 0:
        y_coords = (y_coords / max_y) * 0.5
    else:
        # If max_y is zero or not meaningful, skip
        continue

    # Shift distribution upwards
    ax.fill_between(x_coords, y_coords + i*y_off, i*y_off, alpha=0.7, color=cmap(i/12))
    ax.plot(x_coords, y_coords + i*y_off, color='black', linewidth=1)
    
    # Plot a vertical line for climate metric (MeanTemp)
    # If MeanTemp is on a different scale, consider a second axis or normalization.
    mean_temp = climate_stats.loc[climate_stats['MonthNum'] == month, 'MeanTemp'].values[0]
    # We'll just place a white line at MeanTemp as a reference
    ax.axvline(mean_temp, ymin=(i*y_off)/(12*y_off), ymax=((i*y_off)+0.6)/(12*y_off), 
               color='white', linewidth=2)

    # Add month labels on the left
    ax.text(filtered_df['Percentage'].min() - 2, i*y_off + (y_off/2), 
            calendar.month_abbr[month], va='center', ha='right', fontsize=11)

# Clean up axes
ax.set_yticks([])
ax.set_yticklabels([])
ax.set_ylabel('')
ax.set_xlabel('Referral Percentage (%)')
ax.set_title('Monthly Referral Percentage Distribution (2013-2019) with Climate Overlay')
sns.despine(left=True, bottom=False)

# Adjust limits so we can see labels
ax.set_ylim(-y_off, len(month_order)*y_off)
ax.set_xlim(low_cut, high_cut)

plt.tight_layout()
plt.show()
```

```python
import ee
import pandas as pd

# Initialize Earth Engine
ee.Initialize()

def get_monthly_climate_data(start_date, end_date, geometry):
    # Load ERA5 monthly dataset
    era5 = ee.ImageCollection('ECMWF/ERA5/MONTHLY') \
        .filterDate(start_date, end_date) \
        .select('mean_2m_air_temperature')
    
    # Function to extract mean temperature per month
    def extract_monthly(image):
        stats = image.reduceRegion(
            reducer=ee.Reducer.mean(),
            geometry=geometry,
            scale=30000
        )
        temp = stats.get('mean_2m_air_temperature')
        return ee.Feature(None, {
            'Month': image.date().format('YYYY-MM'),
            'Temperature_K': temp
        })

    monthly_data = era5.map(extract_monthly).getInfo()
    climate_list = [f['properties'] for f in monthly_data['features']]

    climate_df = pd.DataFrame(climate_list)
    climate_df['Month'] = pd.to_datetime(climate_df['Month'])
    # Convert Kelvin to Celsius
    climate_df['Temperature'] = climate_df['Temperature_K'] - 273.15
    climate_df.drop(columns='Temperature_K', inplace=True)

    # Create MonthNum for merging
    climate_df['MonthNum'] = climate_df['Month'].dt.month

    return climate_df

# Example usage:
# Define a geometry. For example, Cape Town:
geometry = ee.Geometry.Point([18.4241, -33.9249])
climate_df = get_monthly_climate_data('2013-01-01', '2019-12-31', geometry)

# climate_df now has columns: ['Month', 'Temperature', 'MonthNum']
print(climate_df.head())
```

```python
import pandas as pd

# Calculate the IQR (interquartile range) for referral percentages
q1 = filtered_df['Percentage'].quantile(0.25)  # 25th percentile (Q1)
q3 = filtered_df['Percentage'].quantile(0.75)  # 75th percentile (Q3)
iqr = q3 - q1  # Interquartile range

# Define lower and upper bounds for outliers
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Filter for outliers
outliers_df = filtered_df[(filtered_df['Percentage'] < lower_bound) | (filtered_df['Percentage'] > upper_bound)]

# Display the months and years corresponding to outliers
outliers_summary = outliers_df[['Month', 'Percentage', 'Temperature']]
outliers_summary['Year'] = outliers_summary['Month'].dt.year
outliers_summary['MonthName'] = outliers_summary['Month'].dt.month_name()

# Sort by year and month for readability
outliers_summary = outliers_summary.sort_values(by=['Year', 'Month'])

# Display the outliers
from IPython.display import display

# Display the outliers
display(outliers_summary)
```

```python
# ---------------------------------------------------------
# Remove Outliers from the Dataset
# ---------------------------------------------------------

# Define the outlier rows (from the table provided)
outlier_indices = [
    399, 400, 401, 402, 403, 404, 405,  # October 2017
    406, 407, 408, 409, 410, 411, 412,  # November 2017
    413, 414, 415, 416, 417, 418, 419   # December 2017
]

# Drop the outlier rows
filtered_df_no_outliers = filtered_df.drop(index=outlier_indices)

# ---------------------------------------------------------
# Violin Plot: After Removing Outliers
# ---------------------------------------------------------
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize

# Compute mean temperature for color mapping
monthly_stats = filtered_df_no_outliers.groupby('MonthName')['Temperature'].mean().reset_index()
temp_min = monthly_stats['Temperature'].min()
temp_max = monthly_stats['Temperature'].max()

# Normalize temperatures for color mapping
norm = Normalize(vmin=temp_min, vmax=temp_max)
cmap = plt.cm.coolwarm

# Create color mapping for each month
month_color_dict = {
    month: cmap(norm(temp)) for month, temp in zip(monthly_stats['MonthName'], monthly_stats['Temperature'])
}

# Violin Plot
sns.set_theme(style="whitegrid", font_scale=1.2)
plt.figure(figsize=(10, 6))
sns.violinplot(data=filtered_df_no_outliers, x='MonthName', y='Percentage',
               palette=month_color_dict, inner='box', linewidth=1)

plt.title('Violin Plot of Referral Percentage by Month (Excluding Outliers)', fontsize=16, pad=20)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Referral Percentage (%)', fontsize=12)
plt.xticks(rotation=45)

# Add colorbar for temperature mapping
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=plt.gca(), fraction=0.02, pad=0.05)
cbar.set_label('Mean Monthly Temperature (°C)', fontsize=12)

plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# Box Plot: After Removing Outliers
# ---------------------------------------------------------

plt.figure(figsize=(10, 6))
ax = sns.boxplot(data=filtered_df_no_outliers, x='MonthName', y='Percentage',
                 palette=month_color_dict, linewidth=1.2)

plt.title('Box Plot of Referral Percentage by Month (Excluding Outliers)', fontsize=16, pad=20)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Referral Percentage (%)', fontsize=12)
plt.xticks(rotation=45)

# Add colorbar for temperature mapping
cbar = plt.colorbar(sm, ax=ax, fraction=0.02, pad=0.05)
cbar.set_label('Mean Monthly Temperature (°C)', fontsize=12)

plt.tight_layout()
plt.show()
```
